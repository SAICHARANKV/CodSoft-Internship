### 1. Import Required Modules and Data


```python
## required libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

pd.set_option('display.max_columns', None)
sns.set_style('darkgrid')
```

```python
## reading dataset
df = pd.read_csv('creditcard.csv')
print( 'All Modules & Data Imported Sucessfully')
```

### 5. Model Training

Selected `Feature` columns and `Target` columns and trained and splitted the dataset.

```python
## dependent and independent variables
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
## train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

##### As it is a classification model, we tried with some of most prominent Classification algorithms to find which classifier is best.

```python
## Standard Scaler
sc=  StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
lg = LogisticRegression()

## fit on training data
lg.fit(X_train, y_train)
```

```python
rf = RandomForestClassifier()

## fit on training data
rf.fit(X_train, y_train)
```

###  6. Accuracy of Model

```python
## l g prediction
pred_lg = lg.predict(X_test)

print('Classification Report for Linear Regression:  \n', classification_report(y_test, pred_lg))
print("-" * 100)
print()
print('Accuracy Score: ', accuracy_score(y_test, pred_lg))
print("-" * 100)
print()
plt.figure(figsize=(10, 10))
sns.heatmap(confusion_matrix(y_test, pred_lg), annot=True, fmt='g');
```

```python
## prediction
pred_rf= rf.predict(X_test)


print('Classification Report: \n', classification_report(y_test, pred_rf))
print("-" * 100)
print()
print('Accuracy Score: ', accuracy_score(y_test, pred_rf))
print("-" * 100)
print()
plt.figure(figsize=(10, 10))
sns.heatmap(confusion_matrix(y_test, pred_rf), annot=True, fmt='g');
```

### 7.Conclusion

* **Both Classifications seems to be more accurate but Random Forest Classifier shows a slight more accuracy than linear regression.**
* **Both `Heat Maps` shows that data training and test both shows some errors and can be rectified by adjusting test size and with more accurate data.**
